```r
colnames(df)
```

```r
head(df, 10) %>% mutate_if(is.numeric, ~round(., 2))
```

```r
df |>
  filter(price > 250000 & size > 1000) |>
  mutate(ppsqft = price / size) |>
  group_by(neighborhood) |>
  summarise(mean_ppsqft = mean(ppsqft),
        median_ppsqft = median(ppsqft),
        count = n()) |>
  mutate_if(is.numeric, ~round(., 2)) |>
  arrange(neighborhood)
```
