## Part A:  &nbsp;Translation

Integers greater than 4 should be squared.  
Integers equal to or less than 4 should be added to 1.  
y should be [2, 4, 25, 49]  
and the output should be that the mean of y is 20.

```python

x = np.array([1, 3, 5, 7])
y = np.where(x > 4, x**2, x + 1)
print(y)
np.mean(y)
```
